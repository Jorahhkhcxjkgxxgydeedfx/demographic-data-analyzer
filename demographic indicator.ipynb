```python
import pandas as pd
import random

# Sample list of categories for categorical columns
workclass = ['State-gov', 'Self-emp-not-inc', 'Private', 'Federal-gov', 'Local-gov', 'Self-emp-inc', 'Without-pay', 'Never-worked']
education = ['Bachelors', 'HS-grad', '11th', 'Masters', 'Associate', 'Some-college', 'Doctorate', 'Prof-school', '5th-6th', '1st-4th', 'Preschool']
marital_status = ['Never-married', 'Married-civ-spouse', 'Divorced', 'Separated', 'Widowed']
occupation = ['Adm-clerical', 'Exec-managerial', 'Handlers-cleaners', 'Prof-specialty', 'Other-service', 'Sales', 'Tech-support', 'Craft-repair', 'Machine-op-inspct', 'Transport-moving', 'Farming-fishing', 'Priv-house-serv', 'Protective-serv', 'Armed-Forces']
relationship = ['Not-in-family', 'Husband', 'Wife', 'Own-child', 'Unmarried']
race = ['White', 'Black', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo', 'Other']
sex = ['Male', 'Female']
native_country = ['United-States', 'Cuba', 'Jamaica', 'India', 'Mexico', 'South', 'China', 'Germany', 'Philippines', 'Italy', 'Poland', 'Columbia', 'Cambodia', 'Thailand', 'Ecuador', 'Laos', 'Honduras', 'Hungary', 'Guatemala', 'Nicaragua', 'El-Salvador', 'Peru', 'Vietnam', 'Yugoslavia']
salary = ['<=50K', '>50K']

# Generate a sample of 175 records
data = {
    'age': [random.randint(18, 90) for _ in range(175)],
    'workclass': [random.choice(workclass) for _ in range(175)],
    'fnlwgt': [random.randint(10000, 300000) for _ in range(175)],
    'education': [random.choice(education) for _ in range(175)],
    'education-num': [random.randint(1, 16) for _ in range(175)],
    'marital-status': [random.choice(marital_status) for _ in range(175)],
    'occupation': [random.choice(occupation) for _ in range(175)],
    'relationship': [random.choice(relationship) for _ in range(175)],
    'race': [random.choice(race) for _ in range(175)],
    'sex': [random.choice(sex) for _ in range(175)],
    'capital-gain': [random.randint(0, 5000) for _ in range(175)],
    'capital-loss': [random.randint(0, 4000) for _ in range(175)],
    'hours-per-week': [random.randint(20, 60) for _ in range(175)],
    'native-country': [random.choice(native_country) for _ in range(175)],
    'salary': [random.choice(salary) for _ in range(175)]
}

# Create the dataframe
df = pd.DataFrame(data)

# Print a preview of the dataset
print(df.head())

# Optionally, save to a CSV file
df.to_csv("demographic_data.csv", index=False)
```

```python

import pandas as pd

# Load the dataset
df = pd.read_csv("demographic_data.csv")

# Question 1: How many people of each race are represented in this dataset?
race_counts = df["race"].value_counts()

# Question 2: What is the average age of men?
average_age_men = round(df[df["sex"] == "Male"]["age"].mean(), 1)

# Question 3: What is the percentage of people who have a Bachelor's degree?
percentage_bachelors = round((df["education"] == "Bachelors").mean() * 100, 1)

# Question 4: What percentage of people with advanced education make more than 50K?
# Advanced education includes Bachelors, Masters, Doctorate
advanced_education = df["education"].isin(["Bachelors", "Masters", "Doctorate"])
percentage_advanced_education_high_salary = round(
    (df[advanced_education & (df["salary"] == ">50K")].shape[0] / df[advanced_education].shape[0]) * 100, 1
)

# Question 5: What percentage of people without advanced education make more than 50K?
non_advanced_education = ~advanced_education
percentage_non_advanced_education_high_salary = round(
    (df[non_advanced_education & (df["salary"] == ">50K")].shape[0] / df[non_advanced_education].shape[0]) * 100, 1
)

# Question 6: What is the minimum number of hours a person works per week?
min_hours_per_week = df["hours-per-week"].min()

# Question 7: What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?
min_hour_workers = df[df["hours-per-week"] == min_hours_per_week]
percentage_min_hour_workers_high_salary = round(
    (min_hour_workers[min_hour_workers["salary"] == ">50K"].shape[0] / min_hour_workers.shape[0]) * 100, 1
)

# Question 8: What country has the highest percentage of people that earn >50K and what is that percentage?
country_salary_counts = df[df["salary"] == ">50K"]["native-country"].value_counts()
country_counts = df["native-country"].value_counts()
highest_earning_country_percentage = round((country_salary_counts / country_counts * 100).max(), 1)
highest_earning_country = (country_salary_counts / country_counts * 100).idxmax()

# Question 9: Identify the most popular occupation for those who earn >50K in India.
india_high_salary = df[(df["native-country"] == "India") & (df["salary"] == ">50K")]
top_india_occupation = india_high_salary["occupation"].value_counts().idxmax()

# Display the results
print("Race counts:\n", race_counts)
print("Average age of men:", average_age_men)
print("Percentage with Bachelor's degrees:", percentage_bachelors, "%")
print("Percentage with advanced education earning >50K:", percentage_advanced_education_high_salary, "%")
print("Percentage without advanced education earning >50K:", percentage_non_advanced_education_high_salary, "%")
print("Minimum number of hours worked per week:", min_hours_per_week)
print(
    "Percentage of minimum hour workers earning >50K:",
    percentage_min_hour_workers_high_salary,
    "%",
)
print("Country with highest percentage of high earners:", highest_earning_country)
print(
    "Highest percentage of high earners in a country:",
    highest_earning_country_percentage,
    "%",
)
print("Most popular occupation for high earners in India:", top_india_occupation)

```

```python
def demographic_data_analysis():
    import pandas as pd
    df = pd.read_csv("demographic_data.csv")

    # 1. Race counts
    race_counts = df['race'].value_counts()

    # 2. Average age of men
    average_age_men = round(df[df['sex'] == 'Male']['age'].mean(), 1)

    # 3. Percentage with Bachelor's degrees
    percentage_bachelors = round((df[df['education'] == 'Bachelors'].shape[0] / df.shape[0]) * 100, 1)

    # 4. Percentage with advanced education >50K
    advanced = df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
    percentage_high_edu_income = round((advanced[advanced['salary'] == '>50K'].shape[0] / advanced.shape[0]) * 100, 1)

    # 5. Percentage without advanced education >50K
    non_advanced = df[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
    percentage_low_edu_income = round((non_advanced[non_advanced['salary'] == '>50K'].shape[0] / non_advanced.shape[0]) * 100, 1)

    # 6. Minimum hours per week
    min_hours = df['hours-per-week'].min()

    # 7. Percentage of min hours >50K
    min_hours_workers = df[df['hours-per-week'] == min_hours]
    percentage_min_hours_income = round((min_hours_workers[min_hours_workers['salary'] == '>50K'].shape[0] / min_hours_workers.shape[0]) * 100, 1)

    # 8. Country with highest >50K percentage
    country_income = df[df['salary'] == '>50K']['native-country'].value_counts()
    country_population = df['native-country'].value_counts()
    country_percentage = (country_income / country_population) * 100
    highest_earning_country = country_percentage.idxmax()
    highest_earning_country_percentage = round(country_percentage.max(), 1)

    # 9. Most popular occupation in India >50K
    india_high_income = df[(df['native-country'] == 'India') & (df['salary'] == '>50K')]
    top_occupation_india = india_high_income['occupation'].value_counts().idxmax()

    return {
        "race_counts": race_counts,
        "average_age_men": average_age_men,
        "percentage_bachelors": percentage_bachelors,
        "percentage_high_edu_income": percentage_high_edu_income,
        "percentage_low_edu_income": percentage_low_edu_income,
        "min_hours": min_hours,
        "percentage_min_hours_income": percentage_min_hours_income,
        "highest_earning_country": highest_earning_country,
        "highest_earning_country_percentage": highest_earning_country_percentage,
        "top_occupation_india": top_occupation_india
    }

results = demographic_data_analysis()
for key, value in results.items():
    print(f"{key}: {value}")
```
